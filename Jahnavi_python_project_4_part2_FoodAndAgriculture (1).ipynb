# Question 3

```python
# percentage distribution of the data in each year using the pie chart

yearly_data = data['totalprod'].groupby(data['year']).sum()
yearly_data.plot(kind = 'pie', figsize = (10,10), autopct = '%1.1f%%')
plt.title('Percentage of honey production by year')
plt.show()
```

Understanding:
    All the year have almost same amount of distribution which shows that there is even amount of honey production over the years.
    There is a very slight amount of increase in honey production in 2000.
    The production of honey neither increased nor decreased over the years.

# Question 4

```python
# distribution of 'priceperlb'

sns.displot(data['priceperlb'])
plt.title('Price Per Lb vs frequency')
plt.xlabel('Price per lb')
plt.ylabel('Frequency')
plt.show()
```

Understanding:

    Majority of the prices are in the lower range.
    Around 1.5, 1.6 price per lb has higher peaks.

```python
#  relationship between the variables 'numcol' and 'prodval' 
x = data["numcol"]
y = data["prodvalue"]
plt.scatter(x,y)
plt.title("Relationship between Number of Colonies and Production Value")
plt.xlabel("Number of Colonies")
plt.ylabel("Production Value")
plt.show()
```

# Question 6

```python
#  relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot
x = data['year']
y = data['prodvalue']
plt.figure(figsize = (10,10))
sns.boxplot(data = data, x = data['year'], y = data['prodvalue'])
plt.title("Production value vs year")
```
